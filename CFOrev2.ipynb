```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("eStoreData.csv",header=0,sep=",")
```

<h2>This is the main data cleaning and setting up section</h2>

```python
departments=[]
costPerDepartment=[]
for i in range(len(departments)):
    costPerDepartment.append(0)
department_costs = df.groupby('department')['cost'].sum()
```

<h2>The Data prep section</h2>

```python
x=[]
departments = df["department"].unique()
for i in range(len(departments)):
    x.append(departments[i])
    print(x[i])
```

```python
y=[]
for i in range(len(department_costs)):
    y.append(department_costs.iloc[i])

print(x,y)
```

<h2>Data Sorting</h2>

```python
y, x = zip(*sorted(zip(y, x)))
```

<h2>Graphing</h2>

```python
plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.title("Cost Per Department")
plt.xlabel("Departments")
plt.ylabel("Cost")
plt.xticks(rotation=50)
plt.ylim(1.7e6,2.4e6)
plt.show()
```

<h2>IP Filtering</h2>

```python
def cleanIP(ip):
    ip = str(ip)
    val = ip.startswith("0") or ip.startswith("255") or ip.startswith("127") or ip.startswith("192")
    return not val

df = df[df['ip'].apply(cleanIP)]

y=[]
for i in range(len(department_costs)):
    y.append(department_costs.iloc[i])
    
y, x = zip(*sorted(zip(y, x)))
    
plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.title("Cost Per Department")
plt.xlabel("Departments")
plt.ylabel("Cost")
plt.xticks(rotation=50)
plt.ylim(1.7e6,2.4e6)
plt.show()
```

<h2>Student Purchasing Analysis</h2>

```python
def cleanEDU(email):
    val = email.endswith(".edu")
    return not val

df = df[df['email'].apply(cleanEDU)]

y=[]
for i in range(len(department_costs)):
    y.append(department_costs.iloc[i])

y, x = zip(*sorted(zip(y, x)))
    
plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.title("Cost Per Department")
plt.xlabel("Departments")
plt.ylabel("Cost")
plt.xticks(rotation=50)
plt.ylim(1.7e6,2.4e6)
plt.show()
```

```python
visa,mastercard = 0,0
for i in range(len(df['card'])):
    if str(df['card'].iloc[i]).startswith("4"):
        visa += 1
    elif str(df['card'].iloc[i]).startswith("5"):
        mastercard += 1
    elif str(df['card'].iloc[i]).startswith("5"):
        mastercard += 1
print(visa,mastercard)
```

```python
plt.figure(figsize=(10,5))
plt.pie([visa, mastercard], labels=["Visa", "Mastercard"], autopct="%1.1f%%")
plt.show()
```
